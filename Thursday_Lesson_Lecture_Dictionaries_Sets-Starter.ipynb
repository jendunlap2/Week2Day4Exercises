## Set <br>

##### Declaring

##### .union() 

```python
# Returns a union of two sets, can also use '|' or set.union(set)
# joins all numbers, gets rid of duplicates
s1 = {1,2,3,4}
s2 = {3,4,5,6}
```

```python
s3 = s1.union(s2)
print(s3)

# OR

s4 = s2 | s1
print(s4)
```

##### .intersection()

```python
# Returns an intersection of two sets, can also use '&'
# only takes similar elements from both sets


s5 = s1.intersection(s2)
print (s5)

# OR

s6 = s2 & s1
print(s6)
```

##### .difference()

```python
# Returns a set containing all the elements of invoking set that are not in the second set, can also use '-'
# only takes values from the first set that are not in the second set
# order matters

s7 = s1.difference(s2)
print(s7)

# OR

s7 = s1 - s2
print(s7)

s8 = s2.difference(s1)
print(s8)

s8 = s2 - s1
print(s8)
```

### 1) Create a program that allows a user to continue to add people to an address book until the user quits. Once the user quits, break out of the loop and print out the name and address of everyone in the address book

###### Steps
1. Create a function that will ask user for name and addresses and stores them in a dictionary
2. Define an empty dictionary with which to work (global or local variable?)
3. Begin a loop that will continue to ask a user for information until the user "quits"
4. If the user does not quit, ask for a name and address and store the variables into variables
5. Add information to the dictionary with name as the key and address as the value
6. If the user does quit, end the loop
7. Print out the information from the dictionary in a formatted way
8. Execute/Call the function

```python
from IPython.display import clear_output

def make_address_book():
    
    addr_book = {}
    keep_going = True
    
    while keep_going == True:
        stop_or_go = input("Would you like to add an entry? y/n ")
        if stop_or_go.lower() == 'y':
            name = input('Name: ')
            address = input('Address: ')
            addr_book[name.title()] = address

        else:
            keep_going = False
            
    for name, address in addr_book.items():
        print(f"{name.title()} lives on {address.title()}.")
    
make_address_book()

```

### 2) Best Time to Meet

Billy is trying to figure out if there is a time that he and his team can meet to work on the project. His three teammates each give him a list of times they are available ('HH:MM' 24-hour). Create a function that will take in the list of all the teammates and return a list of times where everyone can meet.

```python
billy_times = ['09:00', '10:30', '11:30', '12:00', '13:00', '14:30']
teammate_1 = ['09:30', '10:00', '10:30', '12:00', '14:30', '16:00']
teammate_2 = ['09:00', '09:30', '11:00', '11:30', '12:00', '13:30', '14:30', '15:00']
teammate_3 = ['11:00', '11:30', '12:00', '14:00', '14:30', '16:30', '17:00']
# Available Times: '12:00' and '14:30'

# define the function
def lets_meet(*args):
    
    # give value of first list, turn type to set for intersection method
    final_time = set(args[0])
    
    #loop through each list, intersecting with the previous list as you go
    for i in args:
        final_time = final_time.intersection(set(i))
 
    return list(final_time)


lets_meet(billy_times, teammate_1, teammate_2, teammate_3)

```
